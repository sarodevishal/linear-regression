```python
#numpy:it is a library in python which indicates number.
#it provides high performance multidimentional arrayobject with these array.


#Arrays 
#a numpy array is a grid of value,all of the same type,and is indexed by a tuple of non-negative integers.
#The number of dimentions is the rank of the array,
#the shape of an array is a tuple of integers giving the size of the array along each dimentions



import numpy as np
a=np.array([1,2,3])
print(a)#print the elements in the array
print(type(a))#it shows what is the type of the array
print(a.shape)#it gives size of the array
print(a[0],a[1],a[2])
a[0]=5#updating the element in an list
print(a)

b=np.array([[1,2,3],[4,5,6]])#it gives the size of the array
print(b.shape)
print(b[0,1])# it gives values of the index
print(b[1,1])
print(b[1,2])
print(b[0,1],b[0,2],b[1,1])

```

```python
import numpy as np
import matplotlib.pyplot as plt


#computing the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.tan(x)

#here we are ploting the graph
plt.plot(x,y)
plt.show()
```

```python
#how to create multiple graphs at once
import numpy as np
import matplotlib.pyplot as plt

#ploting the graph with respective sin and cos
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

x_cos=np.cos(x)

#plot the points using matplotlib.pyplot
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.plot('x axis label')
plt.plot('y axis label')
plt.title("graph which represent the sin and cos")#it gives the title of the graph
plt.legend(["sin","cos"])
plt.show()
```

```python
#subplots
import numpy as np
import matplotlib.pyplot as plt

#compute the x and y coordinbates for the points sin and cos
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set up a subplot grid that has a height 2 and width 1
#and set first such subplot as active
plt.subplot(2,1,1,)


#make the first plot
plt.plot(x,y_sin)
plt.title('sin')


#set the second subplot as active ,and make a second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cos')

#showin the plot 
plt.show()
```
